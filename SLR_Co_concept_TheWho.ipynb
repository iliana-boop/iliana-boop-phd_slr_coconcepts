```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# DataFrame with the data
data = {
    "Actor": [
        "Academic",
        "Community (Civil Society)",
        "Organized Civil Society",
        "Government",
        "Private Organizations",
        "NGOs"
    ],
    "Co-design": [12, 11, 7, 4, 2, 3],
    "Co-create": [4, 4, 3, 1, 1, 2],
    "Co-produce": [3, 3, 3, 1, 0, 1],
    "Co-develop": [4, 3, 1, 2, 1, 0]
}

df_new = pd.DataFrame(data)
df_new.set_index("Actor", inplace=True)

# Heatmap from the new data
plt.figure(figsize=(10, 6))
ax = sns.heatmap(df_new, annot=True, fmt="d", cmap="YlGnBu", cbar=True)
plt.title("Frequency of Co-Concept Approaches by Actor")
plt.ylabel("Actor")
plt.xlabel("Co-Concept")
plt.tight_layout()
plt.show()
```
